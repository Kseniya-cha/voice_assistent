```python
import speech_recognition as sr 
import os
import playsound
from gtts import gTTS
import gtts
#import pyttsx3


def stop():
    say('пока-пока')
    exit

def start():
    print('запускаем ассистента...')
    while True:
        command = listen().lower()
        handle_command(command)


# распознавание речи юзера
def listen():
    voice_recognizer = sr.Recognizer()
    with sr.Microphone() as micro:
        # настройка чувствительности микрофона
        voice_recognizer.adjust_for_ambient_noise(micro, duration=1)
        audio = voice_recognizer.listen(micro)       
    try:
        voice_text = voice_recognizer.recognize_google(audio,language='ru')
        print(f'вы сказали: {voice_text}')
        return voice_text
    except sr.UnknownValueError:
        return 'ошибка распознавания'
    except sr.RequestError:
        return 'ошибка запроса'
    return command

# речь бота
def say(text):
    voice = gTTS(text, lang='ru')
    file = 'audio.mp3'
    voice.save(file)
    playsound.playsound(file)
    os.remove(file)
    print(f'ассистент: {text}')


def my_evaluate(eq):
    for i in range(len(eq)):
        if eq[i] == 'х':
            eq = eq[:i] + '*' + eq[i+1:]
    eq = int(eval(eq))
    say(f'Я посчитала и получила: {eq}')


# доступные команды
commands_database = {'привет':['приветики', {'привет'}], 
"я молодец": ["само собой! спасибо за создание", {'молодец','я'}],
'ты молодец': ['как и ты',{'ты','молодец'}],
"всё супер": ["я знаю, шикарно работаю", {'супер'}],
"какая прелесть": ["и не говори", {'прелесть'}],
"ничего не получается": ["ничего страшного, сделай перерыв", {'не','получается'}],
"сложно": ["ты справишься", {'сложно'}],
"надо диплом делать": ["да не надо, программирование ведь круче", {'диплом','делать'}],
'как дела': ['всё хорошо, ведь я существую и развиваюсь', {'как','дела'}]}
val = list(commands_database.values()) # отдельно значения словаря

repeat = {'повторяй за мной',"повтори за мной", 'повтори',"повторяй",'скажи'}

# ответ на команды
def handle_command(command):
    command = command.lower()
    # проверка вхождения в словарь по ключевым словам
    command_str = command.split(' ')
    counter = 0
    for elem_say in command_str:
        for j in range(len(val)):
            if elem_say in val[j][1]:
                counter += 1
                command_real = list(commands_database.keys())[j]
        
    # повторялка, 2 условия из-за разных диапазонов
    if (' '.join((command.split(' '))[:3])) in repeat:
        say(' '.join((command.split(' '))[3:]))
    elif (' '.join((command.split(' '))[:1])) in repeat:
        say(' '.join((command.split(' '))[1:]))

    # вычисление
    elif (' '.join((command.split(' '))[:1])) in ['посчитай', "вычисли"]:
        my_evaluate(' '.join((command.split(' '))[1:]))


    # если в словаре - даёт ответ; идёт не в начале, чтобы реализовывались другие
    elif counter > 0:
        if counter == len(commands_database[command_real][1]):
            say(commands_database[command_real][0])
    
    # прощание
    elif command == 'пока':
        stop()
    
    # не распознаёт
    else:
        say("ничё не понимаю")


# конечная функция
def talk():
    print('общение:')
    try:
        start()
    except KeyboardInterrupt:
        stop()

talk()
```
